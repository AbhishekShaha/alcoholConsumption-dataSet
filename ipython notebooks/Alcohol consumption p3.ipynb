```python
# Now, read new files again and merge them. 
beer = pd.read_csv('/Users/abhishek/Desktop/visualization-project/wip/beer.csv', encoding="ISO-8859-1")
wine = pd.read_csv('/Users/abhishek/Desktop/visualization-project/wip/wine.csv', encoding="ISO-8859-1")
spirits = pd.read_csv('/Users/abhishek/Desktop/visualization-project/wip/spirits.csv', encoding="ISO-8859-1")
other = pd.read_csv('/Users/abhishek/Desktop/visualization-project/wip/other.csv', encoding="ISO-8859-1")
```

```python
mer_data = beer.merge(wine,left_on='name',right_on='name',how='left')
```

```python
mer_data.head()
```

```python
mer1_data=mer_data.merge(spirits,left_on='name',right_on='name',how='left')
```

```python
mer1_data.head()
```

```python
mer2_data = mer1_data.merge(other,left_on='name',right_on='name',how='left')
```

```python
mer2_data.head()
```

```python
mer2_data.to_csv('/Users/abhishek/Desktop/visualization-project/wip/merged_data.csv', encoding="ISO-8859-1")
```
